- 优先快速响应

- 适合实时代码完成和快速迭代

- 通常延迟低，但在复杂任务上的准确性略有降低

- 花更多时间“深入思考”问题

- 擅长复杂调试、架构决策和多步问题解决

- 具有高级推理能力，可逐步解析复杂错误


## 3. 多模态强力型

- 能处理代码、文本，还能处理图片、图表甚至视频内容

- 特别适合理解视觉文档或处理UI/UX元素

- 以DeepSeek为代表

- 提供与闭源模型相当的AI能力

- 无需付费或注册

- 可能缺少图像生成或网页浏览等功能

# 提示词工程

## 明确性和清晰性：撰写有效提示

实例：

> 编写一个Python函数sort_by_lastname(customers)，它接收一个包含客户记录的列表（每个记录有first_name和last_name
> 字段），并返回一个按last_name字母顺序排序的列表。请包含简短的文档字符串，并通过将缺失的last_name视为空字符串来处理这种情况。

这段提示明确了语言（Python）、函数名称及其用途、输入结构、排序键、额外要求（文档字符串）和一个边界情况。它很可能产生正是你所需要的结果或非常接近。基本上
，要像规格编写者那样思考：你对任务的描述越精确，AI就越少需要猜测，所需的修改也就越少。

## 具体性策略大纲


### 1. 提及语言或环境

- 例如：“编写一个JavaScript函数...” 而不是仅说“编写函数...”

- 如果针对特定框架或版本，需说明（如“使用React Hooks...”或“Python 3...”）


### 2. 定义输出范围

- 想要单个函数？完整文件或模块？

- 是否包含测试？

- 不同指令如“仅提供函数实现”和“提供完整可运行脚本”会产生不同结果


### 3. 包含需求和限制

- 例如登录示例中指定密码长度和尝试次数限制

- 考虑边界情况和约束并写入提示

- 若需性能优化或特定算法，需明确说明（如“O(n)时间和O(1)空间”或“使用二分查找法”）


### 4. 避免含糊引用

- 避免使用“它”等不明确的代词

- 例如用“处理数组并返回结果数组”替代“处理它并返回结果”


### 5. 指明期望的输出格式

- 明确要求仅代码、带注释代码或附带解释

- 例如“仅给代码，不解释”或“提供代码并简短注释”





## 避免的做法

### 1. 不要写成小说一样冗长

- 冗长且包含无关信息的提示会使模型困惑或关注错误的内容。

- 描述应简洁但完整。

- 例如，在编码环境下通常不需要前置“你是一名世界级程序员”的说明，这样可能增加噪音。


### 2. 不要假设AI会自动正确填充细节

- 如果某些细节重要（如线程安全、特殊字符处理等），一定要明确提及。

- 未提及时，假设AI可能无法处理。


### 3. 避免开放式“创意”提示，需确定性输出时

- 例如，“写一些代码来分析数据”会让AI猜测需求。

- 应明确指定具体任务，如“计算一组数字的平均值和标准差”。


### 二、总结

- 明确表达你的需求。

- AI了解得越多，结果越准确。

- 如果需要多次修正AI的回答，反思提示是否清晰

## 迭代改进：与 AI 的反馈循环

### 2. 反馈的重要性与过程

- 通过反馈引导 AI 逐步靠近预期结果

- 实时训练 AI 以解决具体问题

- 高级提示工程的循环过程示意（提示 → AI输出 → 审查 → 优化提示 → AI输出 ...）

- 每次迭代改动保持小范围，避免失去已有成果

- 让 AI 诊断问题并提出修复建议

# Python 变量和数据类型

## 变量 variable
Python中的变量是用于存储数据值的命名标识符。变量本质上是内存中的一个位置，它用来保存信息，便于程序在计算过程中访问和操作这些数据。


在Python中，变量具有以下特点：


1. 动态类型：变量不需要声明类型，Python会根据赋值自动推断变量的数据类型。

2. 赋值：使用“=”操作符给变量赋值，例如：`x = 10`，`name = "Alice"`。

3. 可变性：变量的值可以随时改变，可以重新赋予不同类型的数据。


简单示例：

```python

x = 5       # 变量x保存整型值5

name = "Bob" # 变量name保存字符串"Bob"

x = 3.14    # 变量x现在保存浮点数3.14
...

Python命名变量的规范主要包括以下几点：


1. 只能包含字母（A-Z，a-z）、数字（0-9）和下划线（_），且不能以数字开头。

2. 区分大小写，变量名x和X是不同的变量。

3. 变量名不能是Python的关键字或内置函数名。

4. 推荐使用小写字母和下划线的组合形式（称为snake_case），如my_variable，便于阅读。

5. 变量名应具有描述性，能够反映变量的用途，避免使用单字符（非循环计数变量）。

6. 避免使用大写字母开头的变量名，因为在Python中大写开头通常用来表示类名。


例如：


```python

user_name = "Alice"
# 表示用户年龄

user_age = 25


# 商品价格

product_price = 19.99


# 是否激活状态

is_active = True


# 存储列表中的最大值

max_value = 100
```

在Python中，可以使用内置函数 `id()` 来获取变量的内存地址。`id()` 函数返回的是对象的“身份”，通常是对象在内存中的地址（但这取决于具体的
Python实现）。


示例如下：


```python

a = 10

print(id(a))

```


这会输出变量 `a` 所引用的对象在内存中的地址（以整数形式表现）。


如果你想以十六进制形式查看地址，可以这样写：


```python

print(hex(id(a)))

```


这样可以更直观地看到内存地址。


需要注意的是，`id()` 返回的内存地址是对象的唯一标识符，适用于当前Python运行时环境。

# 数据类型

数据类型是编程中用于定义变量或常量所能存储数据的种类和性质的分类。简单来说，数据类型决定了数据的取值范围、占用的存储空间以及可以对数据进行的操作。


常见的数据类型包括：


1. 整型（Integer）：表示整数，如1、-5、100等。

2. 浮点型（Float/Double）：表示带有小数点的数值，如3.14、-0.001等。

3. 字符型（Char）：表示单个字符，如'a'、'中'等。

4. 字符串类型（String）：表示一串字符组成的文本，如"Hello"。

5. 布尔型（Boolean）：表示真（true）或假（false）两种值。

6. 其他复杂类型：如数组、结构体、类等，用来表示更复杂的数据结构。


不同的编程语言对数据类型的支持和具体实现可能会有所不同。数据类型帮助程序员更好地管理数据，提高程序的安全性和效率。

Python 中常见的数据类型及其例子如下：


1. 整数（int）
  
- 例子：`a = 10`


2. 浮点数（float）
  
- 例子：`b = 3.14`


3. 字符串（str）
  
- 例子：`c = "Hello, World!"`


4. 布尔值（bool）
  
- 例子：`d = True` 或 `e = False`


5. 列表（list）
  
- 例子：`f = [1, 2, 3, 4, 5]`


6. 元组（tuple）
  
- 例子：`g = (1, 2, 3)`


7. 集合（set）
  
- 例子：`h = {1, 2, 3, 4}`


8. 字典（dict）
  
- 例子：`i = {"name": "Alice", "age": 25}`

在 Python 中，可以使用内置函数 `type()` 来获取变量或值的数据类型。示例如下：


```python

a = 10

print(type(a))  # 输出：<class 'int'>


b = 3.14

print(type(b))  # 输出：<class 'float'>


c = "Hello"

print(type(c))  # 输出：<class 'str'>


d = [1, 2, 3]

print(type(d))  # 输出：<class 'list'>
...

通过调用 `type()` 并传入变量名或值，Python 会返回该对象对应的数据类型类别。

Prompt:
> 从现在开始，你是我的Python教练。我们来练习Python的数据类型。你将给我出一道关于数据类型的判断题，我来回答，你负责判断我的答案对错。

## 字符串类型

Python 中的字符串（string）是不可变的（immutable），这是设计语言时的一种选择，主要原因包括：


1. **保证安全性和稳定性**  
  
不可变对象在多线程环境下更安全，因为它们的状态不会被意外改变，避免了很多并发问题。


2. **方便哈希与字典键的使用**  
  
不可变对象可以被哈希（hash），因此可以作为字典（dict）或集合（set）的键。如果字符串是可变的，作为键就不稳定，可能导致哈希值改变，引发数据错误。


3. **提高性能和内存效率**  
   
由于字符串不可变，Python 可以对相同字符串对象进行缓存（字符串驻留机制），多个变量可以共享同一个字符串对象，节约内存。


4. **简化实现**  
  
不可变类型的设计让字符串的实现和优化更简单，减少了潜在的错误。


总结来说，字符串不可变是一种权衡设计，既保证了安全性，又提高了性能和使用的灵活性。


Python 中表示字符串（string）类型有多种方法，主要包括以下几种：


1. 单引号（'...'）
  
```python
  
s1 = 'Hello'
  
```


2. 双引号（"..."）
  
```python
  
s2 = "World"
  
```


3. 三重单引号（'''...'''）用于多行字符串
  
...

这些方式都可以用来定义字符串，根据实际需求选择适合的表示方法。


Python中的字符串（string）类型拥有丰富的方法，以下是常用的字符串方法及其用法说明：


1. **capitalize()**
  
- 功能：首字母大写，其他字母小写。
  
- 用法：`str.capitalize()`
  
- 示例：
    
```python
    
"hello world".capitalize()  # 输出 "Hello world"
    
```


2. **casefold()**
  
- 功能：实现更强的字符串小写转换，用于不区分大小写比较。
  
- 用法：`str.casefold()`
  
- 示例：
    
```python
    
"HELLO".casefold()  # 输出 "hello"
    
```


3. **center(width, fillchar=' ')**
  
- 功能：返回一个指定宽度的居中字符串，填充指定字符（默认空格）。
  
- 用法：`str.center(width, fillchar)`
  
- 示例：
    
```python
    
"cat".center(7, '*')  # 输出 "**cat**"
    
```


4. **count(sub[, start[, end]])**
  
- 功能：统计子字符串 `sub` 在字符串中出现的次数，可以指定搜索范围。
  
- 用法：`str.count(sub, start=0, end=len(str))`
  
- 示例：
    
```python
    
"banana".count('a')  # 输出 3
    
```


5. **encode(encoding='utf-8', errors='strict')**
  
- 功能：编码字符串，返回字节类型。
  
- 用法：`str.encode(encoding, errors)`
  
- 示例：
    
```python
    
"hello".encode('utf-8')  # 输出 b'hello'
    
```


6. **endswith(suffix[, start[, end]])**
  
- 功能：判断字符串是否以指定后缀结尾。
  
- 用法：`str.endswith(suffix, start=0, end=len(str))`
  
- 示例：
    
```python
    
"filename.txt".endswith(".txt")  # 输出 True
    
```


7. **expandtabs(tabsize=8)**
  
- 功能：将字符串中的制表符（`\t`）转为空格，空格数由tabsize控制。
  
- 用法：`str.expandtabs(tabsize)`
  
- 示例：
    
```python
    
"a\tb".expandtabs(4)  # 输出 "a   b"
    
```


8. **find(sub[, start[, end]])**
  
- 功能：查找子字符串第一次出现的位置，找不到返回-1。
  
- 用法：`str.find(sub, start=0, end=len(str))`
  
- 示例：
    
```python
    
"hello".find("e")  # 输出 1
    
```


9. **format(*args, **kwargs)**
  
- 功能：格式化字符串。
  
- 用法：`str.format(*args, **kwargs)`
  
- 示例：
    
```python
    
"Hello, {}".format("Alice")  # 输出 "Hello, Alice"
    
```


10. **format_map(mapping)**
   
- 功能：使用映射（字典）格式化字符串。
   
- 用法：`str.format_map(mapping)`
   
- 示例：
     
```python
     
"Hello, {name}".format_map({'name': 'Bob'})  # 输出 "Hello, Bob"
     
```


11. **index(sub[, start[, end]])**
   
- 功能：查找子字符串第一次出现位置，找不到抛出异常ValueError。
   
- 用法：`str.index(sub, start=0, end=len(str))`
   
- 示例：
     
```python
     
"hello".index("e")  # 输出 1
     
```


12. **isalnum()**
   
- 功能：判断字符串是否只包含字母和数字。
   
- 用法：`str.isalnum()`
   
- 示例：
     
```python
     
"abc123".isalnum()  # 输出 True
     
"abc 123".isalnum() # 输出 False
     
```


13. **isalpha()**
   
- 功能：判断字符串是否只包含字母。
   
- 用法：`str.isalpha()`
   
- 示例：
     
```python
     
"abc".isalpha()  # 输出 True
     
"abc123".isalpha() # 输出 False
     
```


14. **isdigit()**
   
- 功能：判断字符串是否只包含数字字符。
   
- 用法：`str.isdigit()`
   
- 示例：
     
```python
     
"123".isdigit()  # 输出 True
     
"12.3".isdigit() # 输出 False
     
```


15. **islower()**
   
- 功能：判断字符串中是否有小写字母且所有字母均为小写。
   
- 用法：`str.islower()`
   
- 示例：
     
```python
     
"hello".islower()  # 输出 True
     
"Hello".islower()  # 输出 False
     
```


16. **isspace()**
   
- 功能：判断字符串是否只包含空白字符。
   
- 用法：`str.isspace()`
   
- 示例：
     
```python
     
" \t\n".isspace()  # 输出 True
     
" abc ".isspace()  # 输出 False
     
```


17. **istitle()**
   
- 功能：判断字符串是否为标题格式（每个单词首字母大写）。
   
- 用法：`str.istitle()`
   
- 示例：
     
```python
     
"Hello World".istitle()  # 输出 True
     
"Hello world".istitle()  # 输出 False
     
```


18. **isupper()**
   
- 功能：判断字符串是否全部为大写字母。
   
- 用法：`str.isupper()`
   
- 示例：
     
```python
     
"HELLO".isupper()  # 输出 True
     
"Hello".isupper()  # 输出 False
     
```


19. **join(iterable)**
   
- 功能：以字符串作为分隔符，将序列中的元素连接成一个新的字符串。
